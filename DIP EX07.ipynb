```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("dip.jpg")
```

```python
cv2.imshow("original",img)
cv2.waitKey(0)
```

```python
grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("gray",grey)
cv2.waitKey(0)
```

```python
plt.title('GRAY IMAGE')
plt.imshow(grey,cmap = 'gray')
```

```python
#remove noise
image = cv2.GaussianBlur(grey,(3,3),0)
```

```python
cv2.imshow("GuassBlur",image)
cv2.waitKey(0)
```

```python
# convolve with proper kernels
sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)  #y
sobelxy = cv2.Sobel(image,cv2.CV_64F,1,1,ksize=5)  #xy
```

```python
cv2.imshow("Sobelx",sobelx)
cv2.imshow("Sobely",sobely)
cv2.imshow("Sobelxy",sobelxy)
cv2.waitKey(0)
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.imshow(grey,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2)
plt.imshow(sobelx,cmap='gray')
plt.title('sobelx')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(sobely,cmap='gray')
plt.title('sobely')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4)
plt.imshow(sobelxy,cmap='gray')
plt.title('sobelxy')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.waitKey(0)
```

```python
laplacian = cv2.Laplacian(grey,cv2.CV_64F)
plt.imshow(laplacian,cmap='gray')
plt.title('laplacian')
plt.show()
```

```python
canny_edges = cv2.Canny(grey, 120, 150)
plt.imshow(canny_edges,cmap='gray')
plt.title('canny_edges')
plt.show()
```
